```python
import seaborn as sns # for data visualization
import pandas as pd # for data analysis
import numpy as np # for numeric calculation
import matplotlib.pyplot as plt # for data visualization
```

# https://indianaiproduction.com/seaborn-heatmap/#4-how-to-create-a-seaborn-heatmap-using-snsheatmap-function

```python
array_2d = np.linspace(1,5,12).reshape(4,3) # create numpy 2D array
 
print(array_2d) # print numpy array
```

```python
sns.heatmap(array_2d) # create heatmap
```

```python
# Load dataset
 
globalWarming_df = pd.read_csv("Who_is_responsible_for_global_warming.csv")
globalWarming_df.head()
```

```python
# set country name as index and drop Country Code, Indicator Name and Indicator Code
 
globalWarming_df  = globalWarming_df.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code'], axis=1).set_index('Country Name')
globalWarming_df
```

```python
# Create heatmap
 
plt.figure(figsize=(16,9))
 
sns.heatmap(globalWarming_df)
```

```python
# vmin and vmax parameter set limit of colormap
 
plt.figure(figsize=(16,9))
 
sns.heatmap(globalWarming_df, vmin = 0, vmax = 21)
```

```python
# change heatmap color using cmap
 
plt.figure(figsize=(16,9))
 
sns.heatmap(globalWarming_df, cmap="coolwarm")
```

```python
# center parameter 
 
plt.figure(figsize=(16,9))
 
sns.heatmap(globalWarming_df, cmap="coolwarm", center = 10.0)
```

```python
# robust
 
plt.figure(figsize=(16,9))
 
sns.heatmap(globalWarming_df, robust = True)
```

```python
# annot (annotate) parameter
 
plt.figure(figsize=(16,9))
 
sns.heatmap(globalWarming_df, annot = True)
```

```python
# pass 2D Numpy array to annot parameter
 
annot_arr = np.arange(1,13).reshape(4,3) # create 2D numpy array with 4 rows and 3 columns
 
sns.heatmap(array_2d, annot= annot_arr)
```

```python
# fmp parameter - add text on heatmap cell
 
annot_arr = np.array([['a00','a01','a02'],
                      ['a10','a11','a12'],
                      ['a20','a21','a22'],
                      ['a30','a31','a32']], dtype = str)
 
sns.heatmap(array_2d, annot = annot_arr, fmt="s") # s -string, d - decimal
```

```python
# annot_kws parameter
 
plt.figure(figsize=(16,9))
 
annot_kws={'fontsize':10, 
           'fontstyle':'italic',  
           'color':"k",
           'alpha':0.6, 
           'rotation':"vertical",
           'verticalalignment':'center',
           'backgroundcolor':'w'}
 
sns.heatmap(globalWarming_df, annot = True, annot_kws= annot_kws)
```

```python
# linewidths parameter - divede each cell of heatmap
 
plt.figure(figsize=(16,9))
 
sns.heatmap(globalWarming_df, linewidths=4)
```

```python
# linecolor parameter - change the color of heatmap line
 
plt.figure(figsize=(16,9))
 
sns.heatmap(globalWarming_df, linewidths=4, linecolor="k")
```

```python
# hide color bar with cbar parameter 
  
plt.figure(figsize=(16,9))
 
sns.heatmap(globalWarming_df, cbar = False)
```

```python
# change style and format of color bar with cbar_kws parameter
   
plt.figure(figsize=(14,14))
 
cbar_kws = {"orientation":"horizontal", 
            "shrink":1,
            'extend':'min', 
            'extendfrac':0.1, 
            "ticks":np.arange(0,22), 
            "drawedges":True,
           }
 
sns.heatmap(globalWarming_df, cbar_kws=cbar_kws)
```

```python
# change the shape of each cell in a square with squre parameter
 
plt.figure(figsize=(16,9))
 
sns.heatmap(globalWarming_df, linewidths = 1, square= True)
```

```python
# change x-axis labels using xticklabels parameter
 
plt.figure(figsize=(16,9))
 
sns.heatmap(globalWarming_df, xticklabels = np.arange(0,15))
```

```python
# change y-axis labels using yticklabels parameter
 
plt.figure(figsize=(16,9))
 
country_code = ['USA', 'GBR', 'IND', 'CHN', 'RUS', 'AUS', 'FRA', 'DEU', 'CAN', 'BRA', 'ARG', 'PAK', 'NPL', 'BGD', 'JPN']
 
sns.heatmap(globalWarming_df, yticklabels = country_code)
```

```python
# create square box 
 
plt.figure(figsize=(10,8))
 
sns.heatmap(globalWarming_df, xticklabels = False, yticklabels = False, cbar=False, linecolor="w", linewidths=1) 
```

```python
# multiple heatmaps using subplots
 
plt.figure(figsize=(30,10))
 
plt.subplot(1,3,1) # first heatmap
sns.heatmap(globalWarming_df,  cbar=False, linecolor="w", linewidths=1) 
 
plt.subplot(1,3,2) # second heatmap
sns.heatmap(globalWarming_df,  cbar=False, linecolor="k", linewidths=1) 
 
plt.subplot(1,3,3) # third heatmap
sns.heatmap(globalWarming_df,  cbar=False, linecolor="y", linewidths=1) 
 
plt.show()
```

```python
# set seaborn heatmap title, x-axis, y-axis label and font size
 
plt.figure(figsize=(16,9))
 
ax = sns.heatmap(globalWarming_df)
 
ax.set(title="Heatmap",
      xlabel="Years",
      ylabel="Country Name",)
 
sns.set(font_scale=2) # set fontsize 2
```

```python
# Heatmap with keyword arguments (kwargs) parameter
 
plt.figure(figsize = (16,9))
 
kwargs = {'alpha':.9,'linewidth':10, 'linestyle':'--', 'rasterized':False, 'edgecolor':'w',  "capstyle":'projecting',}
 
ax = sns.heatmap(globalWarming_df,**kwargs)
```

```python
# sns heatmap correlation
 
plt.figure(figsize=(16,9))
 
sns.heatmap(globalWarming_df.corr(), annot = True)
```

```python
# Upper triangle heatmap
 
plt.figure(figsize=(16,9))
 
corr_mx = globalWarming_df.corr() # correlation matrix
 
matrix = np.tril(corr_mx) # take lower correlation matrix
 
sns.heatmap(corr_mx, mask=matrix)
```

```python
# Lower triangle heatmap
 
plt.figure(figsize=(16,9))
 
corr_mx = globalWarming_df.corr() # correlation matrix
 
matrix = np.triu(corr_mx) # take upper correlation matrix
 
sns.heatmap(corr_mx, mask=matrix)
```
